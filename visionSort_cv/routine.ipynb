```python
# import time
# from espmax import ESPMax
# from BusServo import BusServo
# bus_servo = BusServo()
# arm = ESPMax(bus_servo)
# 
# 
# def calculate_scaling_factor(x=300, y=303, length_proportion=110, breadth_proportion=80, image_center=(1647.5, 1275.5)):
    # factor_x = (image_center[0] - x) / length_proportion
    # factor_y = (image_center[1] - y) / breadth_proportion
    # scaling_factor = (factor_x + factor_y) / 2
    # return scaling_factor
    # 
# def calculate_relative_coordinates(x, y, scaling_factor, image_center=(1647.5, 1275.5)):
    # relative_x = (image_center[0] - x) / scaling_factor
    # relative_y = (image_center[1] - y) / scaling_factor
    # return relative_x, relative_y
# 
# 
# 
# if __name__ == '__main__':
  # arm.go_home() 
  # time.sleep_ms(2000) 
  # (x,y,z) = arm.ORIGIN
  # 
  # new_coord = [300,2248]
  # factor = calculate_scaling_factor()
  # rel_x, rel_y = calculate_relative_coordinates(x=new_coord[0], y=new_coord[1], scaling_factor=factor)
  # arm.set_position((x-rel_x,y-rel_y,z-155),1000)
  # time.sleep_ms(1000) 
  
  
import time
from espmax import ESPMax
from BusServo import BusServo
import time
from SuctionNozzle import SuctionNozzle

bus_servo = BusServo()
arm = ESPMax(bus_servo)
nozzle = SuctionNozzle()
arm.go_home()
time.sleep_ms(2000)
(x, y, z) = arm.ORIGIN

def calculate_scaling_factor(x=300, y=303, length_proportion=110, breadth_proportion=80, image_center=(1647.5, 1275.5)):
    factor_x = (image_center[0] - x) / length_proportion
    factor_y = (image_center[1] - y) / breadth_proportion
    scaling_factor = (factor_x + factor_y) / 2
    return scaling_factor


def calculate_relative_coordinates(x, y, scaling_factor, image_center=(1647.5, 1275.5)):
    relative_x = (image_center[0] - x) / scaling_factor
    relative_y = (image_center[1] - y) / scaling_factor
    return relative_x, relative_y


def move_to_coordinates(coordinates_list):
    
    for new_coord in coordinates_list:
        factor = calculate_scaling_factor()
        rel_x, rel_y = calculate_relative_coordinates(x=new_coord[0], y=new_coord[1], scaling_factor=factor)
  
        arm.set_position((x - rel_x, y - rel_y, z - 130), 2000)
        time.sleep_ms(2000)
        nozzle.on()
        
        # arm.go_home()
        # time.sleep_ms(2000)
        
        arm.set_position((x - rel_x, y - rel_y, z - 50), 2000)
        time.sleep_ms(2000)
        
        arm.set_position((x - 110, y - 82, z - 100), 2000)
        time.sleep_ms(2000)



        # arm.set_position((x - 110, y - 82, z - 160), 2000)
        # time.sleep_ms(2000)
        nozzle.off()
        arm.go_home()
        time.sleep_ms(2000)
        
        
        

if __name__ == '__main__':
    
    coordinates_to_move = [
        [2387, 1468],
        [1425, 1113]
]

    move_to_coordinates(coordinates_to_move)
```
